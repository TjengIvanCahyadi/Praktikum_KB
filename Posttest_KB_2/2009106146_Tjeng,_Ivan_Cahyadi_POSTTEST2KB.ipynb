# **Library**

```python
import pandas as pd
```

# **Baca Berkas CSV**

```python
df = pd.read_csv("5-Minute Crafts.csv")
```

# **Dictionary**
Untuk menyimpan data:
*   Jumlah Record
*   Nama Attribute
*   Jumlah Attribute Tipe Objek
*   Jumlah Attribute Tipe Angka
*   Jumlah Attribute Dengan Nilai Null







```python
dictionary = {
    "Jumlah Record" : 0,
    "Nama Attribute" : [],
    "Jumlah Attribute Tipe Objek" : 0,
    "Jumlah Attribute Tipe Angka" : 0,
    "Jumlah Attribute Dengan Nilai Null" : 0
}
```

# **Mencari Jumlah Record Dataset**

```python
dictionary["Jumlah Record"] = df.shape[0]
```

# **Mencari Semua Nama Attribute pada Dataset**

```python
dictionary["Nama Attribute"] = list(df.columns)
```

# **Mencari Jumlah Attribute Bertipe Objek**

```python
dictionary["Jumlah Attribute Tipe Objek"] = len(df.select_dtypes(include = ["object"]).columns)
```

# **Mencari Jumlah Attribute Bertipe Angka**

```python
dictionary["Jumlah Attribute Tipe Angka"] = len(df.select_dtypes(exclude = ["object"]).columns)
```

# **Mencari Jumlah Attribute dengan Nilai Null**

```python
null_value = 0
for value in dict(df.isnull().sum()).values():
  if (value > 0):
    null_value += value
dictionary["Jumlah Attribute Dengan Nilai Null"] = null_value
```

# **Menampilkan Isi Dictionary**

```python
for key, value in dictionary.items():
  if (len(key) == 13):
    print(f"{key}\t\t\t\t: {value}")
  elif (len(key) == 14):
    print(f"{key}\t\t\t\t:\n")
    for index, i in enumerate(value):
      print(f"{index + 1}. {i}")
    print(f"\n")
  elif (len(key) == 27):
    print(f"{key}\t\t: {value}")
  else:
    print(f"{key}\t: {value}")
```
