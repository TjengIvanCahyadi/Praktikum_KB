# **Import Library yang Dibutuhkan**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Baca Berkas 'Pokemon.csv' Kemudian Simpan Kedalam Varibel**
[Pokemon.csv](https://www.kaggle.com/datasets/abcsds/pokemon)

```python
df = pd.read_csv('Pokemon.csv')
```

# **Pie Chart**


---


*   Grafik Pie 1 dibawah ini menunjukkan persebaran tipe utama elemen pokemon. Terlihat bahwa lumayan banyak pokemon bertipe air dan normal.
*   Grafik Pie 2 dibawah ini menunjukkan perbandingan jumlah pokemon legendaris dan non legendaris. Terlihat bahwa pokemon non legendaris mendominasi.

```python
plt.figure(figsize = (20,20))
plt.pie(df['Type 1'].value_counts(), labels = df['Type 1'].value_counts().keys().to_list(), startangle = 90, autopct = '%.1f%%')
plt.title('Tipe Elemen Utama Pokemon', fontweight='bold')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.pie(df['Legendary'].value_counts(), labels = ['Not Legendary', 'Legendary'], startangle = 90, autopct = '%.1f%%', shadow = True, explode = [0, 0.1], wedgeprops = {'edgecolor' : 'black'})
plt.title('Pokemon Legendaris', fontweight='bold')
plt.legend()
plt.show()
```

# **Bar Chart**


---

Grafik batang dibawah ini menunjukkan jumlah pokemon dari setiap generasi. Dapat dilihat bahwa pokemon dengan generasi 1, 3, dan 5 adalah yang paling banyak.

```python
plt.figure(figsize = (10,10))
plt.title('Generasi Pokemon')
plt.xlabel('Generasi')
plt.ylabel('Jumlah')
plt.bar(df['Generation'].value_counts().keys().to_list(), df['Generation'].value_counts())
plt.show()
```

# **Boxplot**


---

dari boxplot dibawah, dapat terlihat bahwa ada 2 data pencilan speed. Ada pokemon yang memiliki speed > 150 dan ada juga pokemon yang memiliki speed lebih dari 175. Nilai tengah speed pokemon terdapat di antara 50 dan 75.

```python
sns.boxplot(x = df['Speed'])
plt.show()
```
